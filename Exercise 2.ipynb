```python

# Show info of a commit
show_data = repo.git.show("-s", fixing_commit).splitlines()
# show_data = repo.git.show("--help",fixing_commit).splitlines()
for line in show_data:
    print(line)
    
```

```python
# Show the based information
show_data = repo.git.show( "--color-moved-ws=ignore-space-at-eol","-w",  fixing_commit).splitlines()
for line in show_data:
    print(line)
```

```python

#     Show all developers across all branches
commit_authors = repo.git.shortlog("-s", "-n","--all").splitlines()

author_commits = []

for commit_author in commit_authors:
    commit, author = commit_author.split("\t")
    author_commits.append([int(commit), author])
    
    
df = pd.DataFrame(author_commits, columns=['Commit', 'Author'])

df.head()
```

```python
# Top 10 most active developers in the project
sns.barplot(y='Author', x="Commit", data=df[:10])


```

```python
# Code changes between 2 commits
# whitespace = -trailing-space,-indent-with-non-tab,-tab-in-indent
diff_data = repo.git.diff("--stat","--word-diff-regex=[^[:space:]]",fixing_commit + "^", fixing_commit).splitlines()
for line in diff_data:
    print(line)
  
```

```python
# Code changes between 2 commits
diff_data = repo.git.diff("-w ","-c", fixing_commit + "^", fixing_commit).splitlines()
for line in diff_data:
    print(line)
```

```python
# Code changes between 2 commits
diff_data = repo.git.diff("--ignore-space-at-eol", "-b", "-w", "--ignore-blank-lines","-G'(^[^\*# /])|(^#\w)|(^\s+[^\*#/])'",fixing_commit + "^", fixing_commit).splitlines()
for line in diff_data:
    print(line)
```
