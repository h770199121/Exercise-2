```python
# Average number of days between current commit to previous commit
show_data = repo.git.diff( "--name-only","-w",  fixing_commit3,fixing_commit3+"^").splitlines()

time_diff = float(0)
count = float(0)
for line in repo.git.diff("--name-only", fixing_commit3, fixing_commit3+"^").splitlines():
    creation = repo.git.log("-2", "--pretty=%ct","--", line).splitlines()
    count += 1
    if(len(creation)==2):
        time_diff += ((float(creation[0]) - float(creation[1])) / 86400)

avg=time_diff/count
print("%.1f"%avg)
```
